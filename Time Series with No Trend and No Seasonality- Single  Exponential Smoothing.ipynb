```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('MA.csv')
df.head()
```

```python
df.tail()
```

```python
alpha = 0.1
df['Y_pred'] = 0.00
df['Y_pred'][0] = 0.00
df['Y_pred'][1] = df['Y_true'][0]
for i in range(1,39):
    df['Y_pred'][i+1] = alpha * df['Y_true'][i] + (1-alpha) * df['Y_pred'][i]
df.head()                                
```

```python
df.tail()
```

```python
df['Absolute Error'] = 0.00
for i in range(1,39):
    df['Absolute Error'][i] = abs( df['Y_pred'][i] - df['Y_true'][i] )
df.head()                                
```

```python
df.tail()   
```

```python
df['Squared Error'] = 0.00
for i in range(1,39):
    df['Squared Error'][i] = ( df['Y_pred'][i] - df['Y_true'][i] )**2
df.head()                                
```

```python
df.tail() 
```

```python
df['Precent Error1'] = 0.00
for i in range(1,39):
    df['Precent Error1'][i] = abs ((  df['Y_true'][i] - df['Y_pred'][i] ) / df['Y_true'][i])
df.head()                                
```

```python
df.tail() 
```

```python
df['Precent Error2'] = 0.00
for i in range(1,39):
    df['Precent Error2'][i] = ( (  df['Y_pred'][i] - df['Y_true'][i] ) / df['Y_true'][i-1] )**2
df.head()                                
```

```python
df.tail() 
```

```python
df['Precent Error3'] = 0.00
for i in range(1,39):
    df['Precent Error3'][i] = ( (  df['Y_true'][i] - df['Y_true'][i-1] ) / df['Y_true'][i-1] )**2
df.head()                                
```

```python
df.tail() 
```

```python
df['Error'] = 0.00
for i in range(1,39):
    df['Error'][i] = ( df['Y_pred'][i] - df['Y_true'][i] )
df.head()                                
```

```python
df.tail() 
```

```python
df['Squared Error Diff'] = 0.00
for i in range(2,39):
    df['Squared Error Diff'][i] = ( df['Error'][i] - df['Error'][i-1] )**2
df.head()                                
```

```python
df.tail()
```

```python
MSE = np.mean(df['Squared Error'][1:-1])
MSE
```

```python
RMSE = np.sqrt(np.mean(df['Squared Error'][1:-1]))
RMSE
```

```python
MAD = np.mean(df['Absolute Error'][1:-1])
MAD
```

```python
MAPE = np.mean(df['Precent Error1'][1:-1])
MAPE
```

```python
ThielsU = np.sqrt( np.sum( df['Precent Error2'][1:-1] ) / np.sum( df['Precent Error3'][1:-1] ) )
ThielsU 
```
